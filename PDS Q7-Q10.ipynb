```python
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')
shortest_movie = df.loc[df['duration'].idxmin(), 'title']
longest_movie = df.loc[df['duration'].idxmax(), 'title']

print(f"Shortest movie: {shortest_movie}")
print(f"Longest movie: {longest_movie}")
```

```python
df['year'] = pd.to_datetime(df['year'], errors='coerce')
df['imdbRating'] = pd.to_numeric(df['imdbRating'], errors='coerce')
sorted_df = df.sort_values(by=['year', 'imdbRating'], ascending=[True, False])
sorted_df.reset_index(drop=True, inplace=True)

print(sorted_df)
```

```python
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')

# Subset the DataFrame for movies with duration between 30 minutes and 180 minutes
subset_df = df[(df['duration'] >= 1800) & (df['duration'] <= 10800)]

# Reset the index for better readability (optional)
subset_df.reset_index(drop=True, inplace=True)

# Display the subset DataFrame
print(subset_df)
```
