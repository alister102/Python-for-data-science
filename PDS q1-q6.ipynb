```python
import pandas as pd
```

```python
df=pd.read_excel("C:\\Users\\4HIN\\Downloads\\SalesData.xlsx")
df
```

```python
Q1
```

```python
least_sale_per_item=df.groupby('Item')['Sale_amt'].min().reset_index()
least_sale_per_item.columns=['item','Least_sale_amt']
least_sale_per_item
```

```python
Q2
```

```python
df['Year'] = df['OrderDate'].dt.year
total_sales = df.groupby(['Year', 'Region'])['Sale_amt'].sum().reset_index()
total_sales.columns = ['Year', 'Region', 'Total_Sales']
total_sales
```

```python
Q3
```

```python
reference_date = pd.to_datetime('2020-01-01')
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['days_diff'] = (reference_date - df['OrderDate']).dt.days
df[['OrderDate', 'days_diff']].head()
```

```python
Q4
```

```python
manager_salesmen_df = df.groupby('Manager')['SalesMan'].unique().reset_index()
manager_salesmen_df.columns = ['Manager', 'List_of_Salesmen']
manager_salesmen_df
```

```python
Q5
```

```python
region_salesmen_sales = df.groupby('Region').agg(salesmen_count=('SalesMan', pd.Series.nunique),  
total_sales=('Sale_amt', 'sum')).reset_index()
region_salesmen_sales
```

```python
Q6
```

```python
manager_sales = df.groupby('Manager')['Sale_amt'].sum().reset_index()
manager_sales.columns = ['Manager', 'Total_Sales']

total_sales_overall = manager_sales['Total_Sales'].sum()
manager_sales['Percent_Sales'] = (manager_sales['Total_Sales'] / total_sales_overall) * 100

manager_sales[['Manager', 'Percent_Sales']]
```
